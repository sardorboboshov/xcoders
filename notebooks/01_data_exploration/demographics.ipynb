```python
import duckdb
import pandas as pd

from utils import analyze_dataframe
```

```python
df = pd.read_csv('../data/demographics.csv')
```

```python
df["annual_income"] = (
    df["annual_income"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
con = duckdb.connect("../dataset.duckdb")
```

```python
con.execute("CREATE OR REPLACE TABLE demographics AS SELECT * FROM df")
```

```python
new_df = con.execute("SELECT * FROM demographics").fetch_df()
analyze_dataframe(new_df, 'demographics', ['cust_id'])
```

```python
con.close()
```
