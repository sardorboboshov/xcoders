```python
import duckdb
import pandas as pd

from utils import analyze_dataframe
```

```python
df = pd.read_excel('../data/loan_details.xlsx')
```

```python
df["loan_amount"] = (
    df["loan_amount"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
con = duckdb.connect('../dataset.duckdb')
con.execute("CREATE OR REPLACE TABLE loan_details AS SELECT * FROM df")
new_df = con.execute("SELECT * FROM loan_details").fetch_df()
analyze_dataframe(new_df, 'loan_details', ['customer_id'])
con.close()
```
