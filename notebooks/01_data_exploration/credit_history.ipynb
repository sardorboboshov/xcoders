```python
import duckdb
import pandas as pd

from utils import analyze_dataframe
```

```python
df_raw = pd.read_parquet('../data/credit_history.parquet')
```

```python
cols_to_int = ["oldest_credit_line_age", "oldest_account_age_months", "num_delinquencies_2yrs"]
for col in cols_to_int:
    if col in df_raw.columns:
        df_raw[col] = df_raw[col].round(0).astype("Int64")
```

```python
con = duckdb.connect('../dataset.duckdb')
```

```python
con.execute("CREATE OR REPLACE TABLE credit_history AS SELECT * FROM df_raw")
```

```python
test_df = con.execute("SELECT * FROM credit_history").fetch_df()
```

```python
analyze_dataframe(test_df, 'credit_history', ['customer_number'])
```

```python
con.close()
```
